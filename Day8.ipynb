```python
import pandas as pd
```

```python
Data = {
    "Name" : ["A", "B", "C", "D"],
    "Dept": ["IT", "HR", "IT", "HR"],
    "Salary": [50000, 40000, 60000, 45000]
}
df = pd.DataFrame(Data)
df.index= [1,2,3,4]
print(df)
```

```python
# Average salary per department
Avg_salary= df.groupby("Dept")["Salary"].mean()
print(Avg_salary)
```

```python
# Highest paid employee per department
Highest_paid_employee= df["Salary"].idxmax()
print("Highest paid employee is:",Highest_paid_employee)
```

```python
# Count employee per department
emp_count = df.groupby("Dept")["Name"].count()
print(emp_count)
```

```python
#sort department by average salary
sorted_value = Avg_salary.sort_values(ascending=False)
print("Departments sorted by average salary is:", sorted_value)
```
