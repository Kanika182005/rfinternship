```python
import pandas as pd
```

```python
Data = {
    "Name": ["A", "B", "C", "D", "E","F", "G"],
    "Salary": [20000, 30000,60000, 50000, 70000, 80000, 65000],
    "Age" : [20, 21, 25, 26, 36,23, 24]
}
df = pd.DataFrame(Data)
df.index = [1,2,3,4,5,6,7]
print(df)
```

```python
# filter by salary
new_salary = df[df["Salary"]>50000]
print(new_salary)
```

```python
#filter by age
new_age = df[df["Age"]<30]
print(new_age)
```

```python
#combine filter results
filtered_data =df[(df["Salary"]>50000) & (df["Age"]<30)]
print(filtered_data)
```

```python
# Save filtered data to new file
filtered_data.to_csv("filter_data.csv", index=False)
print("filter data saved successfully\n")
```
