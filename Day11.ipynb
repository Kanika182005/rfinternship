```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
Days= np.array(["Mon","Tue","Wed","Thu","Fri"])
Sales= np.array([200,250,300,280,350])

plt.plot(Days,Sales, marker ="o", color= "pink")
plt.title("Sales Trend",fontweight = "bold", fontsize= 20 )
plt.xlabel("Days", fontsize = 16, fontweight= "bold")
plt.ylabel("Sales", fontsize = 16, fontweight= "bold")

# highest and lowest value
highest_sale= np.max(Sales)
lowest_sale= np. min(Sales)
highest_day = Days[np.argmax("Sales")]
lowest_day = Days[np.argmax("Sales")]

# Highlight highest and lowest day
plt.scatter(highest_day, highest_sale+ 10, color= "green", label= "highest sale", s= 150, zorder=5)
plt.scatter(lowest_day, lowest_sale, color= "blue", label= "lowest sale", s=150, zorder =5)

plt.text(highest_day, highest_sale-5 , f'Highest:{highest_sale}', ha='center')
plt.text(lowest_day, lowest_sale +5, f'Lowest:{lowest_sale}', ha='center')

plt.grid(True)
plt.legend()


plt.show()




```
