```python
import matplotlib.pyplot as plt
```

```python
categories= ["Food", "Travel", "Shopping"]
expenses = [500, 300, 200]
plt.pie(expenses,labels= categories , colors=["blue","green","orange"],
         shadow= True,  startangle=90, explode=[0.1,0,0], autopct="%1.1""f%%")
plt.title("Category Breakdown", fontsize=20, fontweight= "bold")
```
