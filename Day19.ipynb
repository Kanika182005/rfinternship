```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python

# Create Dataset

np.random.seed(42)

dates = pd.date_range(
    start="2025-01-01",
    periods=30,
    freq="D"
)

stock_price = np.random.randint(100,200,30)

df = pd.DataFrame({
    "Date": dates,
    "Stock_Price": stock_price
})

print("Stock Dataset")
print(df.head())


# Moving Average (5 Day)

df["Moving_Average"] = (df["Stock_Price"].rolling(window=5).mean())

print("\nDataset with Moving Average")
print(df.head(10))


# Identify Peak Price

peak_price = df["Stock_Price"].max()

peak_day = df.loc[ df["Stock_Price"].idxmax()]

print("\nPeak Stock Price")
print(peak_day)

# Identify Lowest Price (Drop)

lowest_price = df["Stock_Price"].min()

drop_day = df.loc[df["Stock_Price"].idxmin()]

print("\nLowest Stock Price")
print(drop_day)


# Volatility

volatility = df["Stock_Price"].std()

print("\nVolatility =", round(volatility,2))

# Price Trend

plt.figure(figsize=(10,5))

plt.plot(df["Date"], df["Stock_Price"], marker='o')

plt.title("Stock Price Trend")
plt.xlabel("Date")
plt.ylabel("Price")

plt.xticks(rotation=45)

plt.show()

# Price + Moving Average

plt.figure(figsize=(10,5))

plt.plot( df["Date"],df["Stock_Price"], marker='o',label="Stock Price")

plt.plot(df["Date"],df["Moving_Average"],marker='s',label="5-Day Moving Average")

plt.title("Stock Price vs Moving Average")
plt.xlabel("Date")
plt.ylabel("Price")

plt.legend()

plt.xticks(rotation=45)

plt.show()
```
