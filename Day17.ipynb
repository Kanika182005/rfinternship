```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
# create dataset
np.random.seed()
data= {
"Customer_id": range(1,101),
"Age": np.random.randint(18,65,100),
"Spending": np.random.randint(1000,10000,100),
"Visits": np.random.randint(1,20,100)
}

df= pd.DataFrame(data)
print("..............Customer Dataset............")
print(df.head())


#Customer segmentation
def spending_catogory(amount):
    if amount>7000:
        return "High"
    elif amount>4000:
        return "medium"
    else:
        return "low"
df["Catogory"]= df["Spending"].apply(spending_catogory)
print("\n..........Customer Categories.............")
print(df.head())

```

```python
#High value customer
high_value= df[(df["Spending"]>7000) & (df["Visits"]>15)]
print("\n..........High value customers are...............")
print(high_value)

#Low engagement users
low_engage_user= df[df["Visits"]<5]
print("\n.........Low Engagement User..........")
print(low_engage_user)
```

```python
#spending distribution
plt.figure(figsize=(14,6))
sns.histplot(df["Spending"],bins= 10 , kde=True)
plt.title("Spending Distribution", fontsize= 18, fontweight= "bold")
plt.xlabel("Spending", fontsize= 18, fontweight= "bold")
plt.ylabel("Count", fontsize= 18, fontweight= "bold")
plt.show()
```

```python
#Customer categories
category_count= df["Visits"].value_counts()
plt.figure(figsize=(14,6))
bars=plt.bar(category_count.index, category_count.values)

plt.title("Customer Categories", fontsize= 18, fontweight= "bold")
plt.xlabel("Visits", fontsize= 18, fontweight= "bold")
plt.ylabel("Count", fontsize= 18, fontweight= "bold")

for bar in bars:

    y = bar.get_height()

    plt.text(
        bar.get_x() + bar.get_width()/2,
        y,
        int(y),
        ha='center',
        va='bottom'
    )
    
# Business Strategies

print("\nBusiness Strategies")

print("1. High Customers -> Provide premium offers and loyalty rewards.")
print("2. Medium Customers -> Encourage higher spending using discounts.")
print("3. Low Customers -> Attract through marketing campaigns.")
print("4. Low Engagement Users -> Send reminders and personalized offers.")
```
