```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
Months = ["Jan","Feb", "Mar", "Apr", "May","June"]
Sales = [200,300,250,280,220,350]

plt.figure(figsize=(14,10))
plt.tight_layout()

plt.subplot(2,2,1)
plt.plot(Months, Sales, marker="o")
plt.grid(True)
plt.title("Trends of Sales", fontsize=16, fontweight ="bold")
plt.xlabel("Months", fontsize=14, fontweight = "bold")
plt.ylabel("Sales", fontsize=14, fontweight = "bold")
print(" Graph shows the sales trend.")


plt.subplot(2,2,2)
plt.bar(Months, Sales)
plt.grid(True)
plt.title("Comparison of Sales", fontsize=16, fontweight ="bold")
plt.xlabel("Months", fontsize=14, fontweight = "bold")
plt.ylabel("Sales", fontsize=14, fontweight = "bold")
print(" June month has the highest sale.")

plt.subplot(2,2,3)
sns.histplot(Sales, bins=4, kde=True)
plt.grid(True)
plt.title("Distribution of Sales", fontsize=16, fontweight ="bold")
plt.xlabel("Sales", fontsize=14, fontweight = "bold")
plt.ylabel("Count", fontsize=14, fontweight = "bold")
print(" Maximum sale in between 220-240. ")
```
