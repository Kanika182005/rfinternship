```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python

# Create Dataset

data = {
    "Movie_Name": [
        "Movie A", "Movie B", "Movie C", "Movie D",
        "Movie E", "Movie F", "Movie G", "Movie H"
    ],

    "Rating": [
        8.5, 7.2, 9.0, 6.8,
        8.9, 7.5, 9.2, 8.0
    ],

    "Genre": [
        "Action", "Comedy", "Action", "Drama",
        "Drama", "Comedy", "Action", "Drama"
    ],

    "Revenue": [
        500, 300, 700, 250,
        600, 350, 900, 450
    ]
}

df = pd.DataFrame(data)

print("Movie Dataset")
print(df)

# Highest Rated Movie-

highest_rated = df.loc[df["Rating"].idxmax()]

print("\nHighest Rated Movie")
print(highest_rated)

# Most Profitable Genre

genre_revenue = df.groupby("Genre")["Revenue"].sum()

print("\nRevenue by Genre")
print(genre_revenue)

most_profitable = genre_revenue.idxmax()

print("\nMost Profitable Genre:")
print(most_profitable)

# Top 5 Movies


top5 = df.sort_values(
    by="Rating",
    ascending=False
).head(5)

print("\nTop 5 Movies")
print(top5)


# Correlation

correlation = df["Rating"].corr(df["Revenue"])

print("\nCorrelation between Rating and Revenue:")
print(round(correlation,2))

```

```python
# Visualization 1
# Genre vs Revenue

plt.figure(figsize=(6,4))

plt.bar(
    genre_revenue.index,
    genre_revenue.values
)

plt.title("Genre vs Revenue", fontsize=18, fontweight="bold")
plt.xlabel("Genre",fontsize=16, fontweight= "bold")
plt.ylabel("Revenue",fontsize=16, fontweight= "bold")

plt.show()


# Rating Distribution


plt.figure(figsize=(6,4))

sns.histplot(
    df["Rating"],
    bins=5
)

plt.title("Rating Distribution",fontsize=16, fontweight= "bold")
plt.xlabel("Rating",fontsize=16, fontweight= "bold")
plt.ylabel("Frequency",fontsize=16, fontweight= "bold")

plt.show()
```
