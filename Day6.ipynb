```python
import pandas as pd
```

```python


Dataset = {
    "Product": ["A","B","A", "C"],
    "Quantity": [2,1,3,5],
    "Price": [100, 200, 100, 50]
}

df = pd.DataFrame(Dataset)
print(df)
```

```python
# add new column
df["Total"] = df["Quantity"] * df["Price"]
print(df)
```

```python
# total sales per product
sales_per_product= df.groupby("Product")["Total"].sum()
print(sales_per_product)

```

```python
#total revenue
total_revenue = df["Total"].sum()
print("Total is:",total_revenue )
```

```python
# top selling product
top_selling_product= sales_per_product. idxmax()
print(top_selling_product)
```

```python
#sort by revenue
sorted_value= df.sort_values(by="Total")
print(sorted_value)
```
