```python
import matplotlib.pyplot as plt
import numpy as np
```

```python

Students= ["Amit", "Riya", "John"]
Maths = [78,89,90]
Science= [60, 70,97]
English = [56,78,90]
 
x= np.arange(len(Students))
width = 0.25

plt.xticks(x,Students)

plt.bar(x-width, Maths, width=width, label= "Maths")
plt.bar(x, Science, width=width, label= "Science")
plt.bar(x+width, English, width=width, label= "English")

plt.title("Students Performance", fontweight="bold" , fontsize= 20)
plt.xlabel("Subjects", fontweight= "bold", fontsize=16)
plt.ylabel("Marks", fontweight="bold", fontsize=16)

plt.legend()

plt.show()

```
