```python
import pandas as pd
```

```python
Data = {
    "name":["Amit","Riya", "John"],
    "math": [80 , 90, 60],
    "science": [70,88,65],
    "english": [85, 92, 70]
}
df = pd.DataFrame(Data)
df.index= (1,2,3)
print(df)
 
```

```python
#calculate average marks per student
df["Average"] = (df["math"] + df["science"] + df["english"])/3
print(df)
```

```python
# find topper
topper_student= df["Average"].idxmax()
print("topper student is :" , topper_student)
```

```python
#count student above average

Average_value =df["Average"].mean()
print( "Average value is:",Average_value)

count= len(df[df["Average"]>Average_value])
print("Total student above average is:", count)
```

```python
# Add grade column
df["Grade_column"] = ["B", "A","C"]
print(df)
```

```python
# Subject wise average
math_average = df["math"].mean()
print("Maths average is:",math_average)

science_average = df["science"].mean()
print("Science average is:",science_average)

english_average = df["english"].mean()
print("English average is:",english_average)

```
